```python
import scipy.io
from PIL import Image
import numpy as np
import os

lis = os.listdir('../data/douyu_2700/Labels/')

for li in lis:
    if li[-3:]=='mat':
        print li
        label3 = np.ndarray(shape=(640,368,4),dtype=np.float32)
        label = scipy.io.loadmat('../data/douyu_2700/Labels/{}'.format(li))['label']
        label3[:,:,0] = label*100
        label3[:,:,1] = label*100
        label3[:,:,2] = label*100
        label3[:,:,3] = label*255
        im = Image.fromarray(np.uint8(label3))
        #print './douyu_2700/{}_p.png'.format(li)
        im.save('./douyu_2700/Images/{}_p.png'.format(li[:-4]))
```
