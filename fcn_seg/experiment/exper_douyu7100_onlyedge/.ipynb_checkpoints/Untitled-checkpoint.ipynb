```python
import numpy as np
def softmax(x):
    x = np.array(x)
    x = np.exp(x)
    x.astype('float32')

    if x.ndim > 1:
        sumcol = x.sum(axis = 0)
    for row in x:
        for i in range(row.size):
            row[i] = row[i]/float(sumcol[i])
    return x
#测试结果
scores = [[3.0,1.0],[0.2,0.8]]
print label.ndim
print softmax(np.asarray(label))
```
